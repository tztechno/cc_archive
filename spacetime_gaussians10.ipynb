```python
import os
import shutil
import requests
import zipfile

# ---- Configuration ----
SOURCE_URL = "https://huggingface.co/datasets/stpete2/multiview_video/resolve/main/coffee_martini_03.zip"
DEST_PATH = "/content/SpacetimeGaussians/source"
ZIP_PATH = "/content/coffee_martini_03.zip"

# Create the destination folder
os.makedirs(DEST_PATH, exist_ok=True)

# ---- Download ZIP from URL ----
print("Downloading zip file...")
with requests.get(SOURCE_URL, stream=True) as r:
    r.raise_for_status()
    with open(ZIP_PATH, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)

print("Download complete:", ZIP_PATH)

# ---- Extract the ZIP file ----
print("Extracting zip...")
with zipfile.ZipFile(ZIP_PATH, "r") as zip_ref:
    zip_ref.extractall(DEST_PATH)

print("Extraction complete:", DEST_PATH)

# ---- Delete unnecessary ZIP (Optional) ----
os.remove(ZIP_PATH)
print("Zip file removed.")
```
