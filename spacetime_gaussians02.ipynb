```python
# train_colab_safe.py
import argparse
from pathlib import Path
import torch
import os
import sys

# 元コードの utils パス
sys.path.append(".")
from thirdparty.gaussian_splatting.scene import Scene
from thirdparty.gaussian_splatting.helper3dg import get_scene_data
from thirdparty.gaussian_splatting.utils.my_utils import save_model_checkpoint

def train_colab_safe(source_path, model_path, iterations=1000, batch=2, device='cuda'):
    source_path = Path(source_path)
    model_path = Path(model_path)
    model_path.mkdir(parents=True, exist_ok=True)

    print(f"Loading scene from {source_path}")
    # Scene の読み込み（元の Scene クラスが必要）
    scene = Scene(str(source_path))

    optimizer = torch.optim.Adam(scene.model.parameters(), lr=1e-4)

    print("Start training...")
    for it in range(iterations):
        optimizer.zero_grad()
        images, poses = get_scene_data(scene, batch=batch)
        loss = scene.model.compute_loss(images, poses)  # 仮の計算
        loss.backward()
        optimizer.step()

        if it % 100 == 0:
            print(f"Iter {it}/{iterations} - Loss: {loss.item():.6f}")

    # モデル保存
    checkpoint_path = model_path / "model_final.pth"
    save_model_checkpoint(scene.model, checkpoint_path)
    print(f"Finished training. Model saved at {checkpoint_path}")

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument("--source_path", type=str, required=True)
    parser.add_argument("--model_path", type=str, required=True)
    parser.add_argument("--iterations", type=int, default=1000)
    parser.add_argument("--batch", type=int, default=2)
    args = parser.parse_args()

    train_colab_safe(args.source_path, args.model_path, args.iterations, args.batch)
```
